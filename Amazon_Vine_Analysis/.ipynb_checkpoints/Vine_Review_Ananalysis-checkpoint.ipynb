```python
import pandas as pd

df = pd.read_csv("data/vine_table.csv")
```

```python
df.head()
```

```python
#STEP 1
#Filter the data and create a new DataFrame or table to retrieve all the rows where 
#the total_votes count is equal to or greater than 20 to pick reviews that are more
#likely to be helpful and to avoid having division by zero errors later on.

df_votes_20 = df[df['total_votes']>=20]
df_votes_20.head()
```

```python
#STEP 2
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame
#or table to retrieve all the rows where the number of helpful_votes divided by
#total_votes is equal to or greater than 50%.

df_votes_20["temp"] = df_votes_20["helpful_votes"] / df_votes_20["total_votes"]

df_helpvotes_50 = df_votes_20[df_votes_20['temp']>=.50]
del df_votes_20['temp']
del df_helpvotes_50['temp']
df_helpvotes_50.head(20)
```

```python
# STEP3
#Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves
#all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

df_vines_yes = df_helpvotes_50[df_helpvotes_50['vine'] == "Y"]
df_vines_yes.head(10)
```

```python
#STEP 4
#Repeat Step 3, but this time retrieve all the rows where the review was not part of the 
#Vine program (unpaid), vine == 'N'.
df_vines_no = df_helpvotes_50[df_helpvotes_50['vine'] == "N"]
df_vines_no.head(10)
```

```python
#STEP 5
#Determine the total number of reviews, the number of 5-star reviews, and the percentage of 
#5-star reviews for the two types of review (paid vs unpaid).

review_total = len(df_votes_20)
review_5star = len(df_votes_20[df_votes_20['star_rating']==5.0])
review_paid_percent = len(df_vines_yes[df_vines_yes['star_rating']==5.0]) / len(df_votes_20[df_votes_20['vine']=='Y']) * 100
review_not_paid_percent = len(df_vines_no[df_vines_no['star_rating']==5.0]) / len(df_votes_20[df_votes_20['vine']=='N']) * 100

print("Total Reviews: " + str(len(df)))
print("Total Reliable Reviews: " + str(review_total))
print("Total 5 Star Reviews: "+str(review_5star))
print("Total Reliable Paid Reviews: "+str(len(df_votes_20[df_votes_20['vine']=='Y'])))
print("Total Reliable Non-Paid Reviews: "+str(len(df_votes_20[df_votes_20['vine']=='N'])))
print("Percentage of 5 Star Reviews for Paid Reviews: "+str(review_paid_percent))
print("Percentage of 5 Star Reviews for Unpaid Reviews: "+str(review_not_paid_percent))
print("Number of Paid 5 star reviews: "+str(len(df_vines_yes[df_vines_yes['star_rating']==5.0])))
print("Number of Unpaid 5 star reviews: "+str(len(df_vines_no[df_vines_no['star_rating']==5.0])))
print("Number of Paid 1 star reviews: "+str(len(df_vines_yes[df_vines_yes['star_rating']==1.0]) / len(df_votes_20[df_votes_20['vine']=='Y'])* 100))
print("Number of Unpaid 1 star reviews: "+str(len(df_vines_no[df_vines_no['star_rating']==1.0]) / len(df_votes_20[df_votes_20['vine']=='N'])* 100))
```
